```python
# Importing all necessary libraries
import cv2
import glob

directories_pathnames = ['accident_Cropped', 'accident_Raw',
                         'call_Cropped', 'call_Raw',
                         'doctor_Cropped', 'doctor_Raw',
                         'help_Cropped', 'help_Raw',
                         'hot_Cropped', 'hot_Raw',
                         'lose_Cropped', 'lose_Raw',
                         'pain_Cropped', 'pain_Raw', 
                         'thief_Cropped', 'thief_Raw']


pathname = directories_pathnames[1]
train_path = glob.glob("./Assets/" + pathname + "/*.[aA][vV][iI]")
train_path.sort()
all_frames = []
for path in train_path:
    print('Creating for ' + path)
    cam = cv2.VideoCapture(path)
    current_video_frames = []
    while(True):   
        ret,frame = cam.read()
        if ret:
            current_video_frames.append(frame)
        else:
            all_frames.append(current_video_frames)
            break
```

```python
import os
import cv2
import glob
from Katna.image_selector import ImageSelector

image_selector = ImageSelector()
no_of_keyframes = 3

best_frames = []
# Check whether the specified path exists or not
isExist = os.path.exists('./bestframes/'+pathname)
if not isExist:
   os.makedirs('./bestframes/'+pathname)
   print("The new directory is created!")

for index, current_video_frames in enumerate(all_frames):
    print("current index: " + str(index))
    filtered_images = image_selector.select_best_frames(
        current_video_frames, no_of_keyframes)

    cframe = 0
    best_frames.append(filtered_images)
    for img in filtered_images:
        name = './bestframes/' + pathname + '/keyframe_' + \
            str(index) + "_" + str(cframe) + '.jpg'
        cv2.imwrite(name, img)
        cframe += 1
    print("Keyframes Created for: " + str(index) + " Total: " + str(len(filtered_images)))
```
